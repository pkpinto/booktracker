````python
mongodb = MongoClient('localhost')['booktracker']
````

````python
itunes = mongodb['itunes']
content_metadata = mongodb['content_metadata']
assets = mongodb['assets']
````

````python
for assetid in ['432283637',]:

    # itunes
    r = requests.get('https://itunes.apple.com/lookup?id={}&country=gb'.format(assetid))
    rec = r.json()
    rec['_id'] = assetid
    itunes.replace_one({'_id': assetid}, rec, upsert=True)

    # content metadata
    r = requests.get('https://uclient-api.itunes.apple.com/WebObjects/MZStorePlatform.woa/wa/lookup?version=2&id={}&p=mdm-lockup&caller=MDM&platform=omni&cc=gb'.format(assetid))
    rec = r.json()
    rec['_id'] = assetid
    content_metadata.replace_one({'_id': assetid}, rec, upsert=True)

    # details
    artwork_url = rec['results'][assetid]['artwork']['url'].format(
        w=rec['results'][assetid]['artwork']['width'],
        h=rec['results'][assetid]['artwork']['height'],
        f='jpg',
    )
    r = requests.get(artwork_url)
    artwork = io.BytesIO(r.content)

    rec = {
        '_id': assetid,
        'ibook_link': 'ibooks://assetid/{}'.format(assetid),
        'artwork': artwork.getvalue(),
    }
    assets.replace_one({'_id': assetid}, rec, upsert=True)
````
